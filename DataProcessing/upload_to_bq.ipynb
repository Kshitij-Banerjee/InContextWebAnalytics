```python

from google.cloud import bigquery

# Initialize a BigQuery client
client = bigquery.Client()

# Define the dataset and table
dataset_id = "gatech-edu-439206.page_analytics"
table_id = "page_dimension"

# Configure the load job
job_config = bigquery.LoadJobConfig(
    source_format=bigquery.SourceFormat.CSV,
    autodetect=True,  # Automatically infer schema
    skip_leading_rows=1,  # Skip the header row if needed
)

# Load the CSV file
with open("../Data/nodes_export_with_url.csv", "rb") as source_file:
    load_job = client.load_table_from_file(
        source_file, f"{dataset_id}.{table_id}", job_config=job_config
    )

# Wait for the job to complete
load_job.result()

print("Table loaded successfully.")
```
